```
import cv2
import matplotlib.pyplot as plt
#membaca gambar dalam mode grayscale
img = cv2.imread('gambar.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli (Grayscale)')
plt.show()

binary = cv2.threshold(img, 100, 255,  cv2.THRESH_BINARY)[1]
plt.imshow(binary, cmap='gray')
plt.title('Citra Binerisasi)')
plt.show()
```

```
negative = 255 - img
plt.imshow(negative, cmap='gray')
plt.title('Negative Image ')
plt.show()
```

```
import numpy as np

c = 255/ np.log(1 + np.max(img))
log_image = c * (np.log(1 + img))
log_image = np.array(log_image, dtype=np.uint8)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(log_image, cmap='gray')
plt.title('Logarithmic Transformation')
plt.axis('off')
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar dalam mode grayscale
img = cv2.imread('gambar.jpg', cv2.IMREAD_GRAYSCALE)

if img is None:
    print("❌ Gagal membaca gambar. Pastikan file 'gambar.jpg' ada di direktori kerja.")
else:
    # Nilai gamma (atur sesuai kebutuhan)
    gamma = 0.5

    # Transformasi Power-Law (Gamma Correction)
    power_law = np.uint8(255 * (img / 255) ** gamma)

    # Menampilkan hasil
    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.imshow(img, cmap='gray')
    plt.title('Citra Asli')
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.imshow(power_law, cmap='gray')
    plt.title(f'Transformasi Power-Law (γ={gamma})')
    plt.axis('off')

    plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar dalam grayscale
img = cv2.imread('gambar.jpg', cv2.IMREAD_GRAYSCALE)

if img is None:
    print("❌ Gagal membaca gambar. Pastikan file 'gambar.jpg' ada di direktori kerja.")
else:
    # Kernel Sobel X dan Y
    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]])

    sobel_y = np.array([[-1, -2, -1],
                        [0,  0,  0],
                        [1,  2,  1]])

    # Operasi Sobel (konvolusi)
    gx = cv2.filter2D(img, cv2.CV_64F, sobel_x)
    gy = cv2.filter2D(img, cv2.CV_64F, sobel_y)

    # Menggabungkan hasil Sobel X dan Y
    sobel_combined = np.sqrt(gx**2 + gy**2)

    # Normalisasi hasil ke 0–255 agar bisa ditampilkan dengan jelas
    sobel_combined = cv2.normalize(sobel_combined, None, 0, 255, cv2.NORM_MINMAX)
    sobel_combined = np.uint8(sobel_combined)

    # Tampilkan hasil
    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.imshow(img, cmap='gray')
    plt.title('Gambar Asli')
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.imshow(sobel_combined, cmap='gray')
    plt.title('Efek Sobel')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
